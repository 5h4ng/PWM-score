```python
pepxml_file = '../data/YE_20180428_SK_HLA_A0202_3Ips_a50mio_R1_01.pep.xml'
pin_file = '../data/YE_20180428_SK_HLA_A0202_3Ips_a50mio_R1_01.pin'
import numpy as np
np.float_ = np.float64 # to fix a bug in mokapot:`np.float_` was removed in the NumPy 2.0 release. Use `np.float64` instead.
```

```python
import os
import mokapot
import numpy as np
import pandas as pd
import sys
import matplotlib.pyplot as plt

np.random.seed(42)

import logging

# Change to True enable messages and nicely format them:
log = True
if log:
    logging.basicConfig(
        level=logging.INFO,
        format="%(levelname)s: %(message)s",
    )
logging.info("这是一个info级别的日志消息")
```

```python
psms_pin = mokapot.read_pin(pin_file)
#psms_pepxml = mokapot.read_pepxml(pepxml_file, decoy_prefix='DECOY_') # bug: incorrect target decoy psms
```

```python
moka_conf, models_pin = mokapot.brew(psms_pin)
```

```python
moka_conf.psms['Peptide']
```

```python
comet_conf = psms_pin.assign_confidence()
```

```python
comet_conf.levels
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
colors = ("#343131", "#24B8A0")

# Plot the performance:
for ax, level in zip(axs, ['psms', 'peptides']):
    comet_conf.plot_qvalues(level=level, c=colors[0], ax=ax, threshold=0.5,
                           label="Comet best feature")
    moka_conf.plot_qvalues(level=level, c=colors[1], ax=ax,
                           label="mokapot", threshold=0.5)
    ax.legend(frameon=False)

plt.tight_layout()
plt.show()
```

```python
# plot the performance of different test_fdr values
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
colors = ("#343131", "#24B8A0")

moka_conf_list = []
for test_fdr in [0.01, 0.5]:
    moka_conf, models_pin = mokapot.brew(psms_pin, test_fdr=test_fdr, )
    moka_conf_list.append(moka_conf)
    moka_conf.plot_qvalues(level='psms', c=colors[1], ax=axs[0], threshold=0.2,
                           label=f"mokapot (FDR={test_fdr})")
    moka_conf.plot_qvalues(level='peptides', c=colors[1], ax=axs[1], threshold=0.2,
                           label=f"mokapot (FDR={test_fdr})")
```
